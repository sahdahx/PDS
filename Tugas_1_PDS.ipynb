# Activity 2: Bias Spotting

## 1. Health study uses only hospital patients
### a. Potensi bias: Selection bias, sampling bias, Berkson’s bias.
### b. Arah distorsi: Pasien yang sakit parah akan terlalu banyak diwakili (over-represented), sementara orang sehat atau orang sakit yang tidak mampu ke rumah sakit akan kurang diwakili (under-represented).
### c. Strategi mitigasi: Melakukan desain pengambilan sampel berbasis populasi secara umum atau komunitas untuk mencakup individu yang tidak berada di rumah sakit.



## 2. Online survey distributed only via Instagram
### a. Potensi bias: Coverage bias, self-selection bias, non-response bias.
### b. Arah distorsi: Pengguna usia muda dan aktif media sosial akan over-represented, sedangkan kelompok lanjut usia atau masyarakat tanpa akses internet akan under-represented.
### c. Strategi mitigasi: Menggunakan metode survei multi-moda (seperti telepon atau wawancara langsung) untuk menjangkau kelompok yang tidak menggunakan Instagram.




## 3. Crime prediction uses reported crimes as labels
### a. Potensi bias: Measurement bias, reporting bias, historical bias, feedback loop bias.
### b. Arah distorsi: Area dengan patroli polisi yang lebih intensif akan terlihat memiliki tingkat kejahatan yang lebih tinggi (karena lebih banyak yang dilaporkan), sehingga menciptakan distorsi pada profil area tertentu.
### c. Strategi mitigasi: Menggabungkan data laporan polisi dengan survei korban yang tidak melapor untuk mendapatkan gambaran kejahatan yang lebih akurat.

## 1. Give one example each of cross-sectional, time series, and panel data.

### Cross-sectional: Satu survei terhadap berbagai rumah tangga yang dilakukan pada tahun 2026.

## 2. Why can treating panel data as i.i.d. be problematic?

### Menganggap data panel sebagai independent and identically distributed (i.i.d.) adalah masalah karena observasi pada unit yang sama (individu yang sama) di waktu yang berbeda biasanya saling berkorelasi atau memiliki ketergantungan.

### Mengabaikan korelasi dalam unit ini sering kali menyebabkan pengecilan estimasi ketidakpastian (underestimating uncertainty) atau standar error yang salah.



## 3. Define MCAR, MAR, MNAR in one sentence each.

### MCAR (Missing Completely At Random): Data hilang benar-benar secara acak dan tidak bergantung pada nilai variabel apa pun dalam dataset.

### MAR (Missing At Random): Data hilang secara acak namun polanya bergantung pada data lain yang terobservasi dalam dataset.

### MNAR (Missing Not At Random): Data hilang secara tidak acak karena polanya bergantung pada nilai dari variabel yang hilang itu sendiri atau nilai yang tidak terobservasi.



## 4. Name two types of bias and give a real example for each.

### Sampling Bias: Terjadi ketika data tidak representatif terhadap populasi, contohnya survei kesehatan yang hanya mengambil sampel dari pasien di rumah sakit.

### Measurement Bias: Terjadi ketika terdapat perbedaan kualitas pengukuran antar kelompok, contohnya penggunaan timbangan yang tidak akurat atau label kriminalitas yang hanya didasarkan pada laporan polisi di area tertentu.

## a. Bangun data panel “harga saham” sederhana dengan unit id ∈ {A,B,C}, rentang 30 hari (date), dan variabel numerik price.

```python

# Unit id: A, B, C | Rentang: 30 hari | Variabel: price
np.random.seed(42)
ids = ['A', 'B', 'C']
dates = pd.date_range(start="2025-01-01", periods=30)

# Membuat kombinasi id dan date (Long Format)
df_saham = pd.MultiIndex.from_product([ids, dates], names=['id', 'date']).to_frame(index=False)

# Simulasi harga saham (variabel numerik price)
df_saham['price'] = np.random.uniform(100, 200, size=len(df_saham))
df_saham
```

## b. Pastikan data berada pada format long dengan struktur kolom minimal: (id, date, price). Jika data masih wide, lakukan konversi wide → long.

```python
# Struktur: id, date, price
df_saham = df_saham.sort_values(['id', 'date'])
```

## c. Hitung log-return per id: r_t = log(price_t) - log(price_t-1)


```python
df_saham['log_return'] = df_saham.groupby('id')['price'].transform(lambda x: np.log(x).diff())
df_saham['log_return']
```

## d. Bangun dua fitur untuk klasifikasi dini dari setiap seri (per id), misalnya:
### rolling mean 5-hari pada price, dan
### rolling volatility 5-hari (simpangan baku) pada log-return.

```python
# 1. Rolling mean 5-hari pada price
df_saham['rolling_mean_5d'] = df_saham.groupby('id')['price'].transform(lambda x: x.rolling(window=5).mean())

# 2. Rolling volatility 5-hari (SD) pada log-return
df_saham['rolling_vol_5d'] = df_saham.groupby('id')['log_return'].transform(lambda x: x.rolling(window=5).std())

# Menampilkan hasil
df_saham
```
