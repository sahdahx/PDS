## ACTIVITY 2: BIAS SPOTTING

1. **Health study uses only hospital patients**
   
   a. Jenis bias: Selection bias dan sampling bias (termasuk kemungkinan Berkson’s bias).
   
   b. Arah distorsi: Pasien dengan kondisi lebih serius atau yang memiliki akses layanan kesehatan cenderung lebih banyak terwakili, sedangkan individu sehat atau yang sulit mengakses fasilitas kesehatan menjadi kurang terwakili.
   
   c. Strategi mitigasi: Terapkan pengambilan sampel berbasis populasi umum, kombinasikan data rumah sakit dengan survei komunitas, atau gunakan teknik random sampling dari populasi luas.

2. **Online survey distributed only via Instagram**

   a. Jenis bias: Coverage bias dan self-selection bias.

   b. Arah distorsi: Responden yang aktif di Instagram, umumnya lebih muda dan melek teknologi, akan mendominasi data, sementara kelompok yang jarang menggunakan media sosial atau tidak memiliki akses internet menjadi minim representasi.

   c. Strategi mitigasi: Gunakan berbagai kanal distribusi survei, lakukan stratifikasi responden, dan sesuaikan hasil dengan pembobotan agar mendekati struktur populasi sebenarnya.

3. **Crime prediction uses reported crimes as labels**

   a. Jenis bias: Reporting bias dan measurement bias, serta potensi feedback loop bias.

   b. Arah distorsi: Area dengan tingkat pelaporan atau pengawasan tinggi akan tampak lebih rawan, sedangkan wilayah dengan pelaporan rendah bisa terlihat aman meskipun kondisi sebenarnya belum tentu demikian.

   c. Strategi mitigasi: Padukan data laporan resmi dengan survei korban atau sumber komunitas, lakukan evaluasi keadilan model secara rutin, serta gunakan teknik penyesuaian data untuk mengurangi ketimpangan representasi.



# **Quick Check (Quiz-Style Questions)**

1. **Give one example each of cross-sectional, time series, and panel data.**

   Cross-sectional: Survei tingkat literasi digital pada 500 mahasiswa yang dikumpulkan pada satu waktu tertentu, misalnya tahun 2026.

   Time series: Data penjualan bulanan sebuah perusahaan dari Januari 2020 sampai Desember 2025 yang dicatat secara berurutan.

   Panel data: Data 200 mahasiswa yang sama yang diobservasi IPK dan tingkat stresnya setiap semester selama 4 tahun.

2. **Why can treating panel data as i.i.d. be problematic?**

   Menganggap data panel sebagai i.i.d. bermasalah karena pengamatan dari unit yang sama pada waktu berbeda biasanya saling berkaitan. Ketergantungan ini melanggar asumsi independensi dan dapat menyebabkan standar error menjadi tidak akurat.

3. **Define MCAR, MAR, MNAR in one sentence each.**

   MCAR (Missing Completely At Random): Data hilang terjadi sepenuhnya acak dan tidak dipengaruhi oleh variabel mana pun dalam data.

   MAR (Missing At Random): Data hilang berkaitan dengan variabel yang teramati, tetapi tidak bergantung pada nilai yang hilang itu sendiri.

   MNAR (Missing Not At Random): Data hilang dipengaruhi oleh nilai yang tidak teramati atau oleh nilai variabel yang hilang tersebut.

4. **Name two types of bias and give a real example for each.**

   Selection Bias: Terjadi ketika sampel tidak mewakili populasi, misalnya survei politik yang hanya dilakukan pada pengguna media sosial tertentu.

   Measurement Bias: Terjadi ketika alat atau cara pengukuran menghasilkan kesalahan sistematis, misalnya data berat badan yang diukur dengan timbangan yang tidak terkalibrasi dengan baik.

### 3. What are likely data quality issues (missingness, outliers, measurement error)?

- Terkait measurement error, adanya nilai desimal pada variabel yang secara konsep merupakan hitungan (misalnya clicks, add_to_cart, purchase) mengindikasikan kemungkinan proses agregasi, estimasi, atau pembobotan dari sistem platform iklan.
```python
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns

outlier_summary = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    outlier_summary[col] = {
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "jumlah_outlier": outliers.shape[0],
        "persentase_outlier (%)": round(outliers.shape[0] / len(df) * 100, 2)
    }

outlier_df = pd.DataFrame(outlier_summary).T
outlier_df
```

### a. Bangun data panel “harga saham” sederhana dengan unit id ∈ {A,B,C}, rentang 30 hari (date), dan variabel numerik price.

```python
import pandas as pd
import numpy as np

np.random.seed(7)

saham = ['X','Y','Z']
tanggal = pd.date_range("2025-05-01", periods=30)

df = pd.MultiIndex.from_product([saham, tanggal],
                                names=['kode','tanggal']).to_frame(index=False)

df['harga'] = np.random.uniform(80, 180, len(df))
df = df.sort_values(['kode','tanggal']).reset_index(drop=True)
df
```

### b. Pastikan data berada pada format long dengan struktur kolom minimal: (id, date, price). Jika data masih wide, lakukan konversi wide → long.

```python
df_wide = pd.DataFrame({
    'tanggal': tanggal,
    'X': np.random.uniform(80,180,30),
    'Y': np.random.uniform(80,180,30),
    'Z': np.random.uniform(80,180,30)
})

df_long = df_wide.melt(id_vars='tanggal',
                       var_name='kode',
                       value_name='harga')
df_long
```

### c. Hitung log-return per id: r_t = log(price_t) - log(price_t-1)

```python
df['log_ret'] = df.groupby('kode')['harga'].apply(
    lambda x: np.log(x).diff()
).reset_index(level=0, drop=True)
df['log_ret']
```

### d. Bangun dua fitur untuk klasifikasi dini dari setiap seri (per id)

```python
df['mean_3h'] = df.groupby('kode')['harga'].apply(
    lambda x: x.rolling(3).mean()
).reset_index(level=0, drop=True)

df['vol_3h'] = df.groupby('kode')['log_ret'].apply(
    lambda x: x.rolling(3).std()
).reset_index(level=0, drop=True)
```
