```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import expon

np.random.seed(123)
```

## 1. Kasus Distribusi Bernoulli

Kasus:
- Seorang pengguna melihat iklan, dengan peluang klik sebesar 0.3.

Variabel acak:
- X = 1 jika pengguna klik
- X = 0 jika tidak klik

Distribusi:
- X ~ Bernoulli(p = 0.3)

Parameter:
- p = 0.3

```python
p = 0.3
sample = bernoulli.rvs(p, size=1000)
print("Mean empiris:", sample.mean())

plt.hist(sample, bins=[-0.5,0.5,1.5], edgecolor='black')
plt.xticks([0,1])
plt.title("Simulasi Bernoulli (Klik Iklan)")
plt.show()
```

**Interpretasi:**

Distribusi Bernoulli menghasilkan nilai 0 dan 1, dengan mayoritas observasi berada pada nilai 0 (tidak klik) yang sesuai dengan probabilitas sukses yang relatif kecil (p = 0.3). Nilai mean empiris sebesar 0.301 sangat dekat dengan mean teoritis 0.3.

Hal ini menunjukkan bahwa simulasi sudah berjalan dengan baik dan konsisten dengan teori. Perbedaan kecil disebabkan oleh variasi acak dalam simulasi. Secara konsep, ini menunjukkan bahwa rata-rata sampel dari variabel Bernoulli dapat digunakan sebagai estimasi probabilitas kejadian.

## 2. Kasus Distribusi Binomial

Kasus:
- Dari 20 pengguna, peluang masing-masing klik adalah 0.3.

Variabel acak:
- X = jumlah pengguna yang klik

Distribusi:
- X ~ Binomial(n = 20, p = 0.3)

Parameter:
- n = 20
- p = 0.3

```python
n, p = 20, 0.3
sample = binom.rvs(n, p, size=1000)
print("Mean empiris:", sample.mean())

plt.hist(sample, bins=15, edgecolor='black')
plt.title("Simulasi Binomial")
plt.show()
```

**Interpretasi:**

Distribusi Binomial menunjukkan jumlah keberhasilan dari 20 percobaan dengan peluang sukses 0.3. Histogram terlihat membentuk pola mendekati simetris di sekitar nilai tengah.

Mean empiris sebesar 6.108 mendekati mean teoritis np=20×0.3=6. Hal ini menunjukkan bahwa sebagian besar hasil simulasi berada di sekitar 6 keberhasilan.

## 3. Kasus Distribusi Poisson

Kasus:
- Jumlah pelanggan yang datang ke toko per menit rata-rata 4 orang.

Variabel acak:
- X = jumlah pelanggan per menit

Distribusi:
- X ~ Poisson(λ = 4)

Parameter:
- λ = 4

```python
lam = 4
sample = poisson.rvs(lam, size=1000)
print("Mean empiris:", sample.mean())

plt.hist(sample, bins=15, edgecolor='black')
plt.title("Simulasi Poisson")
plt.show()
```

**Interpretasi:**

Distribusi Poisson digunakan untuk memodelkan jumlah kejadian dalam interval waktu tertentu dengan rata-rata tetap λ = 4. Histogram menunjukkan bentuk yang agak miring ke kanan.

Mean empiris sebesar 4.083 sangat dekat dengan mean teoritis λ = 4. Hal ini menunjukkan bahwa simulasi mampu merepresentasikan distribusi teoritis dengan baik. Artinya, model Poisson sesuai untuk kasus jumlah kejadian seperti kedatangan pelanggan per waktu.

## 4. Kasus Distribusi Eksponensial

Kasus:
- Waktu tunggu pelanggan di kasir rata-rata 2 menit.

Variabel acak:
- X = waktu tunggu pelanggan

Distribusi:
- X ~ Eksponensial(λ = 0.5)

Parameter:
- λ = 0.5

```python
lam = 0.5
sample = expon.rvs(scale=1/lam, size=1000)
print("Mean empiris:", sample.mean())

plt.hist(sample, bins=30, edgecolor='black')
plt.title("Simulasi Eksponensial")
plt.show()
```

**Interpretasi:**

Distribusi eksponensial digunakan untuk memodelkan waktu tunggu antar kejadian dengan parameter λ = 0.5. Histogram menunjukkan pola yang jelas miring ke kanan, di mana sebagian besar nilai kecil dan sebagian kecil nilai besar.

Mean empiris sebesar 1.9945 sangat dekat dengan mean teoritis 1/λ=2. Hal ini menunjukkan bahwa waktu tunggu pendek lebih sering terjadi dibandingkan waktu tunggu panjang.

## 5. Eksplorasi LLN

Konsep:
Law of Large Numbers (LLN) menyatakan bahwa rata-rata sampel akan mendekati nilai harapan jika jumlah data besar.

Digunakan pada
Distribusi Bernoulli (klik iklan)

```python
klik = np.random.binomial(n=1, p=0.3, size=5000)
rata_kumulatif = np.cumsum(klik) / np.arange(1, len(klik)+1)

plt.plot(rata_kumulatif)
plt.axhline(0.3, linestyle='--', color='red')
plt.title("LLN pada Bernoulli")
plt.show()
```

**Interpretasi:**

Grafik rata-rata kumulatif pada simulasi Bernoulli menunjukkan bahwa:

- Pada awal simulasi, rata-rata sangat fluktuatif.
- Seiring bertambahnya jumlah observasi, rata-rata semakin stabil.
- Akhirnya, rata-rata mendekati nilai teoritis 0.3.

Hal ini merupakan bukti empiris dari Law of Large Numbers (LLN) yang menyatakan bahwa rata-rata sampel akan mendekati nilai harapan ketika jumlah data semakin besar.

## Latihan 1

Sebuah perusahaan memiliki peluang **0.8** bahwa satu server dapat berjalan normal selama satu hari.

1. Nyatakan variabel acaknya.
2. Distribusi apa yang cocok?
3. Hitung mean dan varians teoritisnya.
4. Simulasikan 1000 hari operasi dengan Python.

```python
p = 0.8
n_sim = 1000

sample = bernoulli.rvs(p, size=n_sim)
print("Mean empiris:", sample.mean())
print("Varians empiris:", sample.var())

# visualisasi
vals, counts = np.unique(sample, return_counts=True)
plt.bar(vals, counts / n_sim, edgecolor='black')
plt.xticks([0,1])
plt.title("Simulasi Bernoulli (1000 hari)")
plt.show()
```

**Interpretasi:**

Soal ini membahas satu percobaan dengan dua kemungkinan, yaitu server normal (sukses) atau server gagal. Karena hanya satu percobaan dengan dua hasil, maka digunakan **Distribusi Bernoulli**. Tujuannya adalah memodelkan status server per hari, memahami mean & varians dari kejadian biner, serta melihat simulasi mendekati teori (LLN).

Misalkan:
- X = 1 → server normal
- X = 0 → server gagal

Hasil simulasi menunjukkan bahwa mean empiris mendekati mean teoritis (0.8).
Hal ini sesuai dengan Law of Large Numbers (LLN), di mana rata-rata sampel akan mendekati nilai harapan saat jumlah data besar. Sebagian besar nilai adalah 1, yang berarti server lebih sering berjalan normal,
sesuai dengan probabilitas tinggi (0.8).

## Latihan 5 — Eksplorasi CLT

Ulangi simulasi CLT untuk ukuran sampel:

- 5
- 20
- 50
- 100

Tuliskan kesimpulan Anda tentang pengaruh ukuran sampel terhadap bentuk distribusi rata-rata sampel.

```python
sample_sizes = [5, 20, 50, 100]
n_sim = 1000

for n in sample_sizes:
    means = []
    for _ in range(n_sim):
        data = np.random.exponential(scale=1, size=n)
        means.append(data.mean())

    plt.hist(means, bins=40, density=True, alpha=0.6)
    plt.title(f"Distribusi mean (n={n})")
    plt.show()
```

**Interpretasi:**

Central Limit Theorem (CLT) pada soal ini digunakan agar rata-rata sampel akan mendekati distribusi normal,
meskipun data asal tidak normal. Tujuannya untuk melihat pengaruh ukuran sampel terhadap bentuk distribusi mean.

Berdasarkan output visualisasi, semakin besar ukuran sampel:

- Distribusi rata-rata semakin mendekati bentuk normal
- Variansi semakin kecil (lebih sempit)
- Estimasi mean semakin stabil

Dapat disimpulkan bahwa ukuran sampel yang lebih besar menghasilkan estimasi yang lebih akurat dan distribusi yang lebih "rapi", sesuai dengan Central Limit Theorem.
